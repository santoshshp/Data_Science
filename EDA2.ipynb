**2. Encoding Techniques**

```python
categorical_features = df.select_dtypes(include=['object']).columns

one_hot_features = [col for col in categorical_features if df[col].nunique() <= 5]
data_one_hot = pd.get_dummies(df[one_hot_features], drop_first=True)
```

```python
from sklearn.preprocessing import LabelEncoder

label_encoded_features = [col for col in categorical_features if df[col].nunique() > 5]
label_encoders = {}

for col in label_encoded_features:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
    label_encoders[col] = le  # Save encoders for reuse.
```

**3. Feature Engineering**

```python
df['age_group'] = pd.cut(df['age'], bins=[0, 25, 50, 75, 100], labels=['Youth', 'Adult', 'Senior', 'Elderly'])
df['income_per_hour'] = df['capital_gain'] / (df['hours_per_week'] + 1)
```

```python
df['fnlwgt_log'] = np.log1p(df['fnlwgt'])  # Apply log transformation to a skewed feature.
```

**4. Feature Selection:**

```python
from sklearn.ensemble import IsolationForest

isolation_forest = IsolationForest(contamination=0.01)  # 1% contamination threshold
outliers = isolation_forest.fit_predict(df[numerical_features])
df = df[outliers == 1]  # Retain only non-outlier rows.
```
